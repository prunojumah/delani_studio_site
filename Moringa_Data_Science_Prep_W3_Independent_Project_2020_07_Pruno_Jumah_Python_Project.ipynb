```
#Determining which of the services offered by MTN is used the most in average
Telcom_dataset = 'Telcom_dataset.csv'

df = pd.read_csv(Telcom_dataset)

#Converting the DATE_TIME column to data type datetime
df['DATETIME']= pd.to_datetime(df['DATETIME']) 

mean_value = df['VALUE'].groupby(df['PRODUTC'])

mean_value.mean()


```

```
#Merging the datasets by columns, 
Telcom_dataset = 'Telcom_dataset.csv'

df = pd.read_csv(Telcom_dataset)

df.rename(columns={'PRODUTC' : 'PRODUCT', 'DATETIME' : 'DATE_TIME'}, inplace=True)

Telcom_dataset2 = 'Telcom_dataset2.csv'

df2 = pd.read_csv(Telcom_dataset2)

df2.rename(columns={'DW_A_NUMBER' : 'DW_A_NUMBER_INT', 'DW_B_NUMBER' : 'DW_B_NUMBER_INT'}, inplace=True)

Telcom_dataset3 = 'Telcom_dataset3.csv'

df3 = pd.read_csv(Telcom_dataset3)

df3.rename(columns={'CELLID' : 'CELL_ID', 'SIET_ID' : 'SITE_ID'}, inplace=True)

df4=pd.concat([df, df2, df3], axis=0)
df4

```

```
#Converting the DATE_TIME column to data type datetime
df4['DATE_TIME']= pd.to_datetime(df['DATE_TIME']) 
df4
```

```
#Determining which of the services offered by MTN is used the most in average

mean_value = df4['VALUE'].groupby(df['PRODUCT'])

mean_value.mean()
```
